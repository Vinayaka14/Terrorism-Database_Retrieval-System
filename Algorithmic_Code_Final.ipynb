```python
# Inverted Indexing
```

```python
# Create an inverted index
inverted_index = defaultdict(set)
for doc_id, doc in enumerate(tokenized_documents):
    for term in doc:
        inverted_index[term].add(doc_id)
```

```python
# Retrieve documents containing a query term
query = 'iran bombing'
result = set()
for term in nltk.word_tokenize(query.lower()):
    if term in inverted_index:
        result |= inverted_index[term]

# Print the matching documents
for doc_id in result:
    print("\n\n\nDOCUMENT: \n"+documents[doc_id])
```

```python
from sklearn.metrics import confusion_matrix

# Convert the retrieved documents to a list
retrieved_docs = [documents[doc_id] for doc_id in result]

# Calculate the confusion matrix
conf_matrix = confusion_matrix(np.array([1 if doc in ground_truth else 0 for doc in documents]),
                               np.array([1 if doc in retrieved_docs else 0 for doc in documents]))

# Calculate precision, recall, and F1-score
tp = conf_matrix[0][0]
fp = conf_matrix[0][1]
fn = conf_matrix[1][0]
precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1_score = 2 * precision * recall / (precision + recall)

# Print the evaluation metrics
print('Confusion matrix:')
print(conf_matrix)
print(f'Precision: {precision:.3f}')
print(f'Recall: {recall:.3f}')
print(f'F1-score: {f1_score:.3f}')
```

```python
# Positional Indexing
```

```python
# Create a positional index
positional_index = defaultdict(dict)
for doc_id, doc in enumerate(tokenized_documents):
    for position, term in enumerate(doc):
        if term not in positional_index:
            positional_index[term] = {}
        if doc_id not in positional_index[term]:
            positional_index[term][doc_id] = []
        positional_index[term][doc_id].append(position)
```

```python
# Tokenize the query
query = 'iran bombing '
query_tokens = nltk.word_tokenize(query.lower())

# Initialize a dictionary to store relevant documents
relevant_docs = defaultdict(list)

# Iterate over query tokens
for i, query_token in enumerate(query_tokens):
    if query_token in positional_index:
        # Iterate over the documents that contain the query token
        for doc_id, positions in positional_index[query_token].items():
            # Iterate over the positions of the query token in the current document
            for pos in positions:
                # Check if the current document already has relevant documents
                if i == 0 or doc_id in relevant_docs[i-1]:
                    relevant_docs[i].append(doc_id)
                    break

a=0
# Print the relevant documents
if relevant_docs:
    doc_ids = set(relevant_docs[len(query_tokens)-1])
    for doc_id in doc_ids:
        print(documents[doc_id])
        a=a+1
else:
    print("No documents found")
```

```python
# Hash-based indexing
```

```python
import hashlib

# Hash indexing the documents
index = {}
for i, doc in enumerate(tokenized_documents):
    for token in doc:
        token_hash = hashlib.sha256(token.encode('utf-8')).hexdigest()  # Generate a unique hash for the token
        if token_hash not in index:
            index[token_hash] = set()
        index[token_hash].add(i)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import AgglomerativeClustering

# Define the number of clusters
num_clusters = 2

# Define the vectorizer to convert tokenized documents into a TF-IDF matrix
vectorizer = TfidfVectorizer()

# Convert tokenized documents to a TF-IDF matrix
tfidf_matrix = vectorizer.fit_transform([' '.join(doc) for doc in d1])

# Perform agglomerative clustering on the TF-IDF matrix
agg_clustering = AgglomerativeClustering(n_clusters=num_clusters, linkage="complete").fit(tfidf_matrix.toarray())

# Get the cluster labels for each document
cluster_labels = agg_clustering.labels_

# Create a dictionary to store the documents for each cluster
cluster_docs = {}

for i in range(num_clusters):
    cluster_indices = np.where(cluster_labels == i)[0]
    cluster_docs[i] = [d2[j] for j in cluster_indices]
```

```python
# Define the query
query = "united states bombing"

# Convert the query to a TF-IDF vector
query_vec = vectorizer.transform([query])

# Check if the query vector has only one sample
if query_vec.shape[0] == 1:
    query_vec = np.append(query_vec.toarray(), np.zeros((1, tfidf_matrix.shape[1])), axis=0)

# Get the cluster labels for the query
query_label = agg_clustering.fit_predict(query_vec)[0]

# Get the documents in the same cluster as the query
cluster_docs = cluster_docs[query_label]

# Print the documents in the same cluster as the query
print("Documents in the same cluster as the query:")
k=0
for doc in cluster_docs:
    print(doc)
    k=k+1
```

```python
from sklearn.metrics import confusion_matrix, classification_report

import re

# Define regular expression patterns to match "united states" and "bombing"
patterns = [re.compile(r'\bunited states\b', flags=re.IGNORECASE), 
            re.compile(r'\bbombing\b', flags=re.IGNORECASE)]

# Define a list to store the matched documents
ground_truth = []

# Loop through each document in d2 and check if it matches both patterns
for doc in d1:
    if all(re.search(pattern, ' '.join(doc)) for pattern in patterns):
        ground_truth.append(doc)

# Define the ground truth labels for the documents
ground_truth_labels = []
for doc in d1:
    if doc in ground_truth:
        ground_truth_labels.append(1)
    else:
        ground_truth_labels.append(0)

# Define the predicted labels for the documents
predicted_labels = []
for doc in d1:
    if doc in cluster_docs:
        predicted_labels.append(1)
    else:
        predicted_labels.append(0)

# Calculate the confusion matrix
cm = confusion_matrix(ground_truth_labels, predicted_labels)

# Calculate the precision, recall, and F1 score
report = classification_report(ground_truth_labels, predicted_labels)

print("Confusion Matrix:")
print(cm)

print("Classification Report:")
print(report)
```
